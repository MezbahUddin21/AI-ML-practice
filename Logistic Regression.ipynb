```python
class LR:

  def __init__(self, learning_rate=0.0001, epochs=1000000):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.weight = None


  def sigmoid(self, z):
     return 1/(1+np.exp(-z))

  def fit(self, x, y):

    self.weight = np.zeros(x.shape[1])

    for epoch in range(self.epochs):
      z = np.dot(x, self.weight)
      pred = self.sigmoid(z)


      grad = np.dot(x.T, y-pred)/len(x)

      self.weight = self.weight + self.learning_rate*grad

      if(epoch%1000==0):
        loss = - np.mean(y*np.log(pred)+(1-y)*np.log(1-pred))
        print(f"At epoch = {epoch} loss is {loss}")


  def weights(self):
    return self.weight


  def predict(self, x_test):

    pred = self.sigmoid(np.dot(x_test, self.weight))
    return np.round(pred)
```

```python
mymodel = LR()
```

```python
mymodel.fit(X_train, Y_train)
```

```python
print(mymodel.weights())
```

```python
Y_pred = mymodel.predict(X_test)
```

```python
accuracy_score(Y_test, Y_pred)
```
